### The Solution: Application Inference Profiles

**Application Inference Profiles (AIP)** solve the multi-tenant problem by providing:

âœ… **Unique model identifiers per tenant** - Each tenant gets their own AIP ARN
âœ… **Isolated CloudWatch metrics** - Separate tracking for each tenant  
âœ… **Accurate cost allocation** - Know exactly what each tenant costs
âœ… **Usage-based billing** - Implement fair pricing with confidence
âœ… **Per-tenant SLA monitoring** - Track performance separately

### How It Works

Instead of all tenants sharing one System Inference Profile:

```python
# âŒ OLD WAY: System Inference Profile (shared by all)
modelId = "us.amazon.nova-pro-v1:0"
```

Each tenant gets their own Application Inference Profile:

```python
# âœ… NEW WAY: Application Inference Profiles (one per tenant)
tenant_a_modelId = "arn:aws:bedrock:us-west-2:123456789012:application-inference-profile/marketing-ai-tenant-a"
tenant_b_modelId = "arn:aws:bedrock:us-west-2:123456789012:application-inference-profile/marketing-ai-tenant-b"
```

**The Magic:** Both AIPs route to the **same underlying model** (Nova Pro 1.0), but CloudWatch tracks them **separately**!

## Section 1: Setup and Initialize AWS Clients

First, let's set up our AWS SDK (boto3) clients for working with Bedrock Application Inference Profiles.

**What we're initializing:**
- `bedrock_client`: For AIP management (create, read, update, delete)
- `bedrock_runtime`: For model inference (we'll use this in Lab 03)
- Configuration from lab helpers (Region, ModelId, Tags)

````python
import boto3
from datetime import datetime
import sys
from lab_helpers.config import Region, ModelId, tenant_a_tags, tenant_b_tags
````

````python
# Initialize Bedrock client
bedrock_client = boto3.client('bedrock', region_name=Region)
bedrock_runtime = boto3.client('bedrock-runtime', region_name=Region)
````

### ğŸ” Get AWS Account ID

We need the AWS account ID to construct proper ARNs for Application Inference Profiles.

**ARN Format:**
```
arn:aws:bedrock:{region}:{account-id}:application-inference-profile/{profile-name}
```

````python
# Get account ID using STS
sts_client = boto3.client('sts')
account_id = sts_client.get_caller_identity()['Account']
print(f"Account ID: {account_id}")
````

````python
# Configuration for the inference profile for Team A (Tenant A)
profile_name = f"marketing-ai-tenant-a-{datetime.now().strftime('%Y%m%d-%H%M%S')}"
description = "Marketing AI AIP for Tenant A"

# Model configuration
MODEL_ID = ModelId
REGION = Region

# Construct proper ARN
MODEL_ARN = f"arn:aws:bedrock:{REGION}:{account_id}:inference-profile/{MODEL_ID}"
print(f"Model ARN: {MODEL_ARN}")

# Model configuration
model_source = {'copyFrom': MODEL_ARN}

# Tags for Tenant A (same as Lab 03)
tags = [
    {'key': 'TenantId', 'value': 'tenant-a'},
    {'key': 'BusinessType', 'value': 'B2B-Tech'},
    {'key': 'Environment', 'value': 'production'},
    {'key': 'CostCenter', 'value': 'marketing-ai-platform'}
]
````

## Section 2: Create Application Inference Profile (Tenant A)

### ğŸ“ Configure AIP for Tenant A

Let's set up the configuration for Tenant A's Application Inference Profile using the **Marketing AI** context:

**Tenant A Profile:**
- **Business Context**: B2B Tech company - SaaS DevOps platform marketing
- **Profile Name**: `marketing-ai-tenant-a-{timestamp}` (unique identifier)
- **Description**: Marketing AI AIP for Tenant A
- **Tags**: Metadata for cost allocation and organization

**Configuration Elements:**
1. **Profile Name**: Unique identifier with timestamp (avoids duplicates)
2. **Description**: Human-readable description of the profile
3. **Model Source**: The System Inference Profile to copy from
4. **Tags**: Metadata for cost tracking and billing

**Tags for Tenant A:**
```python
tags = [
    {'key': 'TenantId', 'value': 'tenant-a'},
    {'key': 'BusinessType', 'value': 'B2B-Tech'},
    {'key': 'Environment', 'value': 'production'},
    {'key': 'CostCenter', 'value': 'marketing-ai-platform'}
]
```

**Why These Tags Matter:**
- **TenantId**: Identifies the specific tenant for billing
- **BusinessType**: Categorizes tenant by business model (B2B vs B2C)
- **Environment**: Separates dev/staging/prod costs
- **CostCenter**: Enables financial reporting and chargebacks

**The `modelSource` Parameter:**
```python
modelSource = {'copyFrom': MODEL_ARN}
```

This tells AWS Bedrock:
- Create a new Application Inference Profile
- Route it to the specified System Inference Profile (Nova Pro 1.0)
- Give it a unique ARN for isolated tracking

````python
# Create the application inference profile for TeamA
try:
    response = bedrock_client.create_inference_profile(
        inferenceProfileName=profile_name,
        description=description,
        modelSource=model_source,
        tags=tags
    )
    print("âœ… Application Inference Profile created successfully for TeamA!")
    print(f"Profile ARN: {response['inferenceProfileArn']}")
    print(f"Profile Name: {profile_name}")
except Exception as e:
    print(f"âŒ Error creating profile: {str(e)}")
````

### ğŸ¯ Create the Application Inference Profile

This is the **critical API call** that creates the Application Inference Profile!

**The API Call:**
```python
bedrock_client.create_inference_profile(
    inferenceProfileName=profile_name,
    description=description,
    modelSource=model_source,
    tags=tags
)
```

**What happens:**
1. AWS Bedrock creates a new Application Inference Profile
2. Assigns it a unique ARN: `arn:aws:bedrock:region:account:application-inference-profile/{name}`
3. Routes it to the specified System Inference Profile (Nova Pro 1.0)
4. Applies the tags for cost tracking
5. Creates a new CloudWatch metrics dimension using the AIP ARN

**Response:**
- `inferenceProfileArn`: The unique ARN for this tenant/team
- `status`: Should be `ACTIVE` when ready

**The Result:** Team A now has their own isolated model endpoint!

## Section 3: Read - Display AIP Tags

### ğŸ·ï¸ Retrieve and Display Tags

Tags are critical for multi-tenant management. Let's retrieve and display the tags associated with Team A's AIP.

**Why Read Tags:**
- Verify the AIP was created with correct metadata
- Audit configuration for compliance
- Validate cost allocation setup
- Document tenant/team mappings

**The API Call:**
```python
bedrock_client.list_tags_for_resource(resourceARN=aip_arn)
```

This returns all tags attached to the Application Inference Profile.

````python
# Display tags for the created profile
if 'response' in locals() and 'inferenceProfileArn' in response:
    try:
        tags_response = bedrock_client.list_tags_for_resource(
            resourceARN=response['inferenceProfileArn']
        )
        
        profile_tags = tags_response.get('tags', [])
        
        print(f"\nğŸ·ï¸ Tags for profile '{profile_name}':")
        if profile_tags:
            for tag in profile_tags:
                print(f"   {tag['key']}: {tag['value']}")
        else:
            print("   No tags found")
            
    except Exception as e:
        print(f"âŒ Error retrieving tags: {str(e)}")
else:
    print("âš ï¸ No profile created to display tags for")
````

## Section 4: Update - Modify AIP Tags

### âœï¸ Update Tags for Tenant A

Let's update the tags for Tenant A's AIP to demonstrate lifecycle management.

**Common Tag Update Scenarios:**
- Tenant upgrades to a higher tier (change Environment tag)
- Team reorganization (update Team or CostCenter)
- Compliance requirements (add new tracking tags)
- Project transitions (update ownership metadata)

**The API Call:**
```python
bedrock_client.tag_resource(
    resourceARN=aip_arn,
    tags=updated_tags
)
```

**Important:** This **adds or updates** tags. Existing tags are merged with new ones. To remove tags, you'd use `untag_resource()`.

**What We're Changing:**

The cell below will update Tenant A's tags with changes from `teamA_tags` in config:
- **Environment**: `production` â†’ `development` (simulating environment change)
- **New Tag Added**: `UpdatedAt: 2025-01-16` (audit trail)

This demonstrates a common scenario where a tenant's project transitions between environments.

````python
# Update tags for Tenant A profile
if 'response' in locals() and 'inferenceProfileArn' in response:
    try:
        # Display current tags before update
        print("ğŸ“‹ Current tags:")
        current_tags_response = bedrock_client.list_tags_for_resource(
            resourceARN=response['inferenceProfileArn']
        )
        for tag in current_tags_response.get('tags', []):
            print(f"   {tag['key']}: {tag['value']}")
        
        # Update with new tags
        print(f"\nğŸ”„ Updating with new tags...")
        print(f"   Updated tags from config:")
        for tag in tenant_a_tags:
            print(f"   {tag['key']}: {tag['value']}")
        
        bedrock_client.tag_resource(
            resourceARN=response['inferenceProfileArn'],
            tags=tenant_a_tags
        )
        
        print(f"\nâœ… Tags updated successfully!")
        
        # Display updated tags
        print(f"\nğŸ“‹ Updated tags:")
        updated_tags_response = bedrock_client.list_tags_for_resource(
            resourceARN=response['inferenceProfileArn']
        )
        for tag in updated_tags_response.get('tags', []):
            print(f"   {tag['key']}: {tag['value']}")
            
    except Exception as e:
        print(f"âŒ Error updating tags: {str(e)}")
else:
    print("âš ï¸ No profile available to update tags for")
````

## Section 5: Clean Up - Delete Both Application Inference Profiles

### ğŸ—‘ï¸ Delete AIPs for Tenant A and Tenant B

After practicing CRUD operations, let's clean up both Application Inference Profiles.

**The API Call:**
```python
bedrock_client.delete_inference_profile(
    inferenceProfileIdentifier=aip_arn
)
```

**What happens:**
1. The Application Inference Profile is marked for deletion
2. No new inference requests can use this AIP
3. Historical CloudWatch metrics remain available
4. Tags and metadata are preserved for audit purposes

**Important Considerations:**
- Deletion is **asynchronous** - takes a few seconds
- Ensure no active inference requests are using these AIPs
- CloudWatch metrics history is **not deleted**
- Consider archiving usage data before deletion for billing records

**When to Delete AIPs:**
- Tenant cancels service (churn)
- Project/team is decommissioned
- Testing/dev AIPs no longer needed
- Consolidating tenants onto shared AIPs

**Note:** Run this cell only after you've completed the Tenant B practice exercise!

````python
# Delete BOTH inference profiles (Tenant A and Tenant B)
# Only run this after completing the Tenant B exercise!

profiles_to_delete = []

# Add Tenant A profile if it exists
if 'response' in locals() and 'inferenceProfileArn' in response:
    profiles_to_delete.append({
        'arn': response['inferenceProfileArn'],
        'name': profile_name,
        'tenant': 'Tenant A'
    })

# Add Tenant B profile if it exists (you should create a tenant_b_response variable)
# if 'tenant_b_response' in locals() and 'inferenceProfileArn' in tenant_b_response:
#     profiles_to_delete.append({
#         'arn': tenant_b_response['inferenceProfileArn'],
#         'name': tenant_b_profile_name,
#         'tenant': 'Tenant B'
#     })

print(f"ğŸ—‘ï¸ Deleting {len(profiles_to_delete)} Application Inference Profile(s)...\n")

for profile in profiles_to_delete:
    try:
        bedrock_client.delete_inference_profile(
            inferenceProfileIdentifier=profile['arn']
        )
        print(f"âœ… {profile['tenant']}: '{profile['name']}' deleted successfully")
    except Exception as e:
        print(f"âŒ {profile['tenant']}: Error deleting '{profile['name']}': {str(e)}")

if len(profiles_to_delete) == 0:
    print("âš ï¸ No profiles available to delete")
    print("   Make sure you've created both Tenant A and Tenant B AIPs first!")
````
